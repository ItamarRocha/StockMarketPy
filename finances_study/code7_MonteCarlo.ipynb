# Predicting Gross Profit

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m,rev_stdev,iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```
